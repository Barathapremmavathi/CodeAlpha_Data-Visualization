```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create images folder
os.makedirs("images", exist_ok=True)

# Load dataset
df = pd.read_csv("Superstore.csv", encoding="ISO-8859-1")

df.head()
df.info()
df.isnull().sum()

# Cleaning
df.dropna(inplace=True)

print("\n")

# 1. Sales Trend
df.groupby("Order Date")["Sales"].sum().plot(figsize=(10,5))
plt.title("Sales Trend Over Time")

plt.savefig("images/sales_trend.png", dpi=300, bbox_inches='tight')  # SAVE IMAGE
plt.show()

print("\n")

# 2. Sales by Category
sns.barplot(x="Category", y="Sales", data=df)
plt.title("Sales by Category")

plt.savefig("images/sales_by_category.png", dpi=300, bbox_inches='tight')
plt.show()

print("\n")

# 3. Profit by Category
sns.barplot(x="Category", y="Profit", data=df)
plt.title("Profit by Category")

plt.savefig("images/profit_by_category.png", dpi=300, bbox_inches='tight')
plt.show()

print("\n")

# 4. Profit Distribution
df.groupby("Category")["Profit"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Profit Distribution")
plt.ylabel("")

plt.savefig("images/profit_distribution.png", dpi=300, bbox_inches='tight')
plt.show()

print("\n")

# 5. Top Products
top = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
top.plot(kind="barh", figsize=(10,6))
plt.title("Top 10 Products")

plt.savefig("images/top_products.png", dpi=300, bbox_inches='tight')
plt.show()

# 6. Region Sales
sns.barplot(x="Region", y="Sales", data=df)
plt.title("Sales by Region")

plt.savefig("images/region_sales.png", dpi=300, bbox_inches='tight')
plt.show()
```
